```python
from pykrx import stock
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from datetime import datetime
import time
import random
```

```python
start_date = "20200101"
end_date = "20230101"
```

```python
ticker_list = stock.get_market_ticker_list()
sellected_stocks = []
while len(sellected_stocks) < 30:
    chosen_stock = random.choice(ticker_list)
    if chosen_stock in sellected_stocks: continue
    # 과한 비중 부여 방지를 위해 평균 시가총액 1000억 ~ 1조 사이에서만 추출
    try:
        평균시가총액 = stock.get_market_cap(start_date, end_date, chosen_stock, "y")['시가총액'].mean()
    except:
        print(f"{chosen_stock} has no C                       AP")
        continue
    if 평균시가총액 < 100_000_000_000 or 평균시가총액 > 1_000_000_000_000: continue
    
    sellected_stocks.append(chosen_stock)
    time.sleep(0.5)
sellected_stocks
```

```python
df = pd.DataFrame()
for ticker in sellected_stocks:
    PRB = stock.get_market_fundamental(start_date, end_date, ticker)["PBR"]
    종목 = stock.get_market_ticker_name(ticker)

    PRB.name = 종목
    df = PRB if df.empty \
        else pd.concat([df, PRB], axis=1)
    time.sleep(0.5)
df.head()
```
